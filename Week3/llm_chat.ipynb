```python
from openai import OpenAI
```

```python
model_1b = 'llama3.2:1b'

base_url = "http://localhost:11434/v1"

ollama = OpenAI(base_url=base_url, api_key="ollama")
```

```python
system_message = "You are helpful assistant"

system_mes = [{'role':'system', 'content':system_message}]
```

```python
messages = [
    {'role':'system', 'content':system_message},
    {'role':'user', 'content': 'My name is John. Count number of letters in my name.'},
    {'role':'system', 'content': 'You have 4 letters in your name - j, o, h, n'},
    {'role':'user', 'content': 'What is my name?'}
]
```

```python
stream = ollama.chat.completions.create(
    model = model_1b,
    messages= messages,
    stream= True
)
```

```python
full_text = ''

for chunk in stream:
    full_text += chunk.choices[0].delta.content or ''

print(full_text)
```
