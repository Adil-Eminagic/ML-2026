```python
print(df["Age"].mean()) # avaegae value, sum of all values / number of itmes
print(df["Age"].median()) # in sorted arry by given value, postion in the middle id even, if odd mean of two items in the middle
print(df["Age"].mode()) # most repeated value, useful for categorcal data
print(df["Age"].std())  # square root of variance
print(df["Age"].var()) # measure of spread of data points around their mean (average) value
```

```python
print(df['MTRANS'].mode())
print(df['CALC'].mode())
```

```python
print(df['MTRANS'].unique())
print(df['CALC'].unique())
```

```python
#column range - difference between the biggest and the lowest value 

print(df["Weight"].max())
print(df["Weight"].min())

print(df["Weight"].max() - df["Weight"].min())
```

**Visualization**

```python
plt.hist(df['Age'])
plt.title("Age distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
plt.hist(df['Weight'])
plt.title("Weight distribution")
plt.xlabel("Weight")
plt.ylabel("Count")
plt.show()
```

```python
df['NObeyesdad'].value_counts().plot(kind='bar')
plt.title("Obesity Level Distribution")
plt.xlabel("Obesity Category")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
sns.scatterplot(data=df, x='Height', y='Weight', hue='NObeyesdad')
plt.title("Height vs Weight by Obesity Level")
plt.show()
```
