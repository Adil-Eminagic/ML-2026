```python
import requests
```

```python
model1 = "mxbai-embed-large:335m" 
model2 = "jeffh/intfloat-multilingual-e5-large-instruct:f32"
```

```python
def get_embedding(text, model=model2):
    url = "http://localhost:11434/api/embeddings"
    payload = {
        "model": model,
        "prompt": text
    }

    response = requests.post(url, json=payload)
    response.raise_for_status()

    data = response.json()
    return data.get("embedding")
```

```python
placeholder_text =  """In a world where technology evolves at an unprecedented pace, individuals and organizations
    are increasingly required to adapt their strategies, workflows, and expectations to keep up
    with emerging innovations. Whether exploring advancements in artificial intelligence, 
    distributed systems, or human-computer interaction, the need for continuous learning has 
    become essential. This environment rewards curiosity, experimentation, and collaboration, 
    encouraging people not only to acquire new technical skills but also to refine their ability 
    to think critically and creatively. As systems grow more interconnected and data flows more 
    rapidly, the ability to interpret information, recognize patterns, and apply insights in 
    meaningful ways becomes central to problem-solving across all domains. The momentum of change 
    brings both new challenges and exciting opportunities, empowering individuals to shape tools, 
    design processes, and build solutions that have real impact on the world around them."""
```

```python
embedding = get_embedding(placeholder_text)
print(f"Embedding vector length: {len(embedding)}")
embedding[:10]  # show first 10 values
```

```python
text = "Hello, world!"
embedding = get_embedding(text)

print("Text: ", text)
print(f"\nEmbedding vector length: {len(embedding)}")
embedding[:10]

```
